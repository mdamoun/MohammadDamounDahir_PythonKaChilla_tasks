# How to create Bar Plot

```python
#Step1: Import the graphs libraries

import seaborn as sns
import matplotlib.pyplot as plt

#Step2: Load Dataset
phool = sns.load_dataset("iris")
phool


#Drarw a line plot
sns.barplot(x="species", y="sepal_width", data=phool)
plt.title("Phoolon ka Plot") #<< Note the Title comes before the command to plot the graph
sns.set_style(style=None, rc=None)
sns.set_style("dark")        #<< mention style before the plot command
plt.show()
```

## Setting Order

replacing male plot position with female and vice versa

```python
#Titanic Data for Grouping
#Step1: Import the graphs libraries

import seaborn as sns
import matplotlib.pyplot as plt

#Step2: Load Dataset
kashti = sns.load_dataset("titanic")
kashti


#Drarw a line plot
sns.barplot(x="sex", y="alone", hue="who", data=kashti, order=["female", "male"]) #< order to be mentioned here
plt.title("Kashti ka Plot") #<< Note the Title comes before the command to plot the graph
sns.set_style(style=None, rc=None)
sns.set_style("dark")        #<< mention style before the plot command
plt.show()
```

## Changing Color of the bars to certain shades of color

## Removing Error bars from the bar chart

## Using Built in Color Palettes for the Chart

```python
#Titanic Data for Grouping
#Step1: Import the graphs libraries

import seaborn as sns
import matplotlib.pyplot as plt

#Step2: Load Dataset
kashti = sns.load_dataset("titanic")
kashti


#Drarw a line plot
sns.barplot(x="sex", y="alone", hue="who", data=kashti, order=["female", "male"], ci = None,
           palette='pastel') #< Defining color palettes to be mentioned here
plt.title("Kashti ka Plot") #<< Note the Title comes before the command to plot the graph
sns.set_style(style=None, rc=None)
sns.set_style("dark")        #<< mention style before the plot command
plt.show()
```

```python
#import the required library
import seaborn as sns
import matplotlib.pyplot as plt

#read titanic data file installed with Seaborn
kashti = sns.load_dataset("titanic")

# facecolor(R,B,G,Alpha) R,G,B, and Aplha range is between Float 0 - 1 so is for all the other values
#facecolor=(R,B,G,Alpha) where RBG is Red, Blue, and Green and Alpha is the degree of transparency.
#The values range between float ( 0 -> 1). Rest for errcolor for CI and edgecolor for the bar is greyscale
#value from black (0) -> white(1)
sns.barplot(x="class", y="fare", data=kashti,
           linewidth=2.5, facecolor=(.7,.5,.5,1),
           errcolor="0", edgecolor="0")

#sns.set_style(style=None, rc=None)
sns.set_style("dark") 
plt.show()
```
