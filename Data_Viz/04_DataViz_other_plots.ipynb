```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

dot = sns.load_dataset("dots")

#defining a color palette

p = sns.color_palette("rocket_r")

#plot line plot

sns.relplot(
            data=dot,
            x="time",
            y="firing_rate",
            hue="coherence",
            size="choice", 
            col="align",
            kind="line",
            size_order=["T1", "T2"],
            palette = p,
            height=5,
            aspect=.75,
            facet_kws=dict(sharex=False)
            )

plt.show()
```

## Nested Plot

```python
#Draw Nested Plots


import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

tips = sns.load_dataset("tips")

#Draw nested plots
sns.boxplot(x="day", y="total_bill",
            hue="smoker", palette=["grey", "white"],
            data=tips)
sns.despine(offset=5, trim=True)  #< Breaking the axises of the plot in to two
```

## Violin Plot / Leaf Plot

```python
#Violin Plot

#Library
import seaborn as sns

#Theme
sns.set_theme(style="whitegrid")


#data
tips = sns.load_dataset("tips")

#Draw Violin Plot
sns.violinplot(x="day", y="total_bill",
            hue="smoker", split=True,
            inner="quart", linewidth=1,
            data=tips, palette={"Yes":"b", "No":".9"})
sns.despine(left=True) 
```

## Scatter Plot

```python
import seaborn as sns
import matplotlib.pyplot as plt

#Theme
sns.set_theme(style="whitegrid")

#data
diamond = sns.load_dataset("diamonds")

#Draw a scatter plot while assigning point colors and sizes to different
#variables in the dataset

f, ax = plt.subplots(figsize=(6.5,6.5))
sns.despine(f, left=True, bottom=True)
clarity_ranking = ["I1", "SI2", "SI1", "VS2", "VS1", "VVS2", "VVS1", "IF"]
sns.scatterplot(x="carat", y="price",
               hue="clarity", size="depth",
               palette="ch:r=-.2, d=.3_r",
               hue_order=clarity_ranking,
               sizes=(1, 8), linewidth=0,
               data=diamond, ax=ax)
```

## Double Plot

```python
import seaborn as sns
import matplotlib.pyplot as plt

#Theme
sns.set_theme(style="ticks")

#data for planets
planets = sns.load_dataset("planets")

#intilaize the figure with logarithmatic x-axis
f, ax = plt.subplots(figsize=(7,6.))
ax.set_xscale("log")

#Plot the orbital period with horizontal boxes
sns.boxplot(x="distance", y="method", data=planets,
           whis = [0, 100], width=.6, palette="vlag")

#Add in point to show each observation
sns.stripplot(x="distance", y="method", data=planets,
             size=4, color=".3", linewidth=0)

#Tweak the visual presentation
ax.xaxis.grid(True)
ax.set(ylabel="")
sns.despine(trim=True, left=True)
```

```python
import seaborn as sns
sns.set_theme(style="darkgrid")

# Load the example Titanic dataset
df = sns.load_dataset("titanic")

# Make a custom palette with gendered colors
pal = dict(male="#6495ED", female="#F08080")

# Show the survival probability as a function of age and sex
g = sns.lmplot(x="age", y="survived", col="sex", hue="sex", data=df,
               palette=pal, y_jitter=.02, logistic=True, truncate=False)
g.set(xlim=(0, 80), ylim=(-.05, 1.05))
```
