# How to create Box and Whisker Plot

```python
#import Seaborn library
import seaborn as sns
#import matplotlib.pyplot as plt

#instruction to chose the style
sns.set(style='whitegrid')

#load the data of fmri
fmri = sns.load_dataset("fmri")

#define the data column to be picked for the boxplot
sns.boxplot(x="timepoint",
           y="signal",
           data=fmri)

#plt.show()
```

```python
#import Seaborn library
import seaborn as sns
import matplotlib.pyplot as plt
import numpy

#instruction to chose the style
sns.set(style='whitegrid')

#load the data of titanic
tips = sns.load_dataset("tips")
tips

#define the data column to be picked for the boxplot
sns.boxplot(x=tips['day'], y=tips['tip'], hue=tips['smoker'], orient='v',
            palette ={"Yes": "red", "No":"green"})

#plt.show()
```

```python
#import the libraries

import seaborn as sns
import pandas as pd
import numpy as np

# load dataset

kashti = sns.load_dataset('titanic')

kashti.head()
```

```python
#adding mean to the Box Plot
sns.boxplot(x="survived", showmeans=True, #<<< Shows the mean
           y="age",
           data = kashti)
```

```python
#changing the mean marker
sns.boxplot(x="survived", showmeans=True, #<<< Shows the mean
            meanprops={"marker":"*", "markersize":"16", "markerfacecolor":"white", "markeredgecolor":None},  #<<< changing the marker
           y="age",
           data = kashti)
```

```python
#Show Labels
sns.boxplot(x="alive", showmeans=True, #<<< Shows the mean
            meanprops={"marker":"*",
                       "markersize":"16",
                       "markerfacecolor":"white",
                       "markeredgecolor":None},  #<<< changing the marker
           hue="sex",
           y="age",
           palette ={"male":"skyblue", "female":"pink"},
           data = kashti)
#Show Lables      #<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<

plt.title("Titanic Age Vs Rate of Sruvival", size="16", weight="bold")
plt.xlabel("Survived")
plt.ylabel("Age of the Passengers (yrs)")
plt.legend()

plt.show()
```
