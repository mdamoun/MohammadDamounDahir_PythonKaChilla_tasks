# Import Libraries

## - Seaborn
Following libraries are installed with Seaborn for it to function. They are automatically installed.
- *NumPy*
- *SciPy*
- *Pandas*
- *Matplotlib*



```python
#Step1: Import the graphs libraries

import seaborn as sns
import matplotlib.pyplot as plt

#Step2: Load Dataset
phool = sns.load_dataset("iris")
phool


#Drarw a line plot
sns.lineplot(x="sepal_length", y="sepal_width", data=phool)
plt.show()
```

## Adding Limits
x and y limits

```python
#Step1: Import the graphs libraries

import seaborn as sns
import matplotlib.pyplot as plt

#Step2: Load Dataset
phool = sns.load_dataset("iris")
phool


#Drarw a line plot
sns.lineplot(x="sepal_length", y="sepal_width", data=phool)
plt.title("Phoolon ka Plot") #<< Note the Title comes before the command to plot the graph
plt.xlim(2) #< Setting x-axis to start from 2
plt.ylim(1) #< Setting y-axis to start from 1
plt.show()

```

## Set Style of the Line Graph
- darkgrid
- whitegrid
- dark
- white
- ticks

```python
#Step1: Import the graphs libraries

import seaborn as sns
import matplotlib.pyplot as plt

#Step2: Load Dataset
phool = sns.load_dataset("iris")
phool


#Drarw a line plot
sns.lineplot(x="sepal_length", y="sepal_width", data=phool)
plt.title("Phoolon ka Plot") #<< Note the Title comes before the command to plot the graph
sns.set_style(style=None, rc=None) #<< This will follow the above style selection command
sns.set_style("dark")        #<< mention style before the plot command
# sns.set_style(style=None, rc=None) #<< This will follow the above style selection command
plt.show()

```

## Size of figure

```python
#Step1: Import the graphs libraries

import seaborn as sns
import matplotlib.pyplot as plt

#Step2: Load Dataset
phool = sns.load_dataset("iris")

#Change Figure
plt.figure(figsize=(8,6))

#Drarw a line plot
sns.lineplot(x="sepal_length", y="sepal_width", data=phool)
plt.title("Phoolon ka Plot") #<< Note the Title comes before the command to plot the graph

plt.show()


```
